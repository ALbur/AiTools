```python
import requests
import random
import string
import os

def generate_response(api_key, prompt, api_url):
    url = f"{api_url}?key={api_key}"
    headers = {
        "Content-Type": "application/json"
    }
    data = {
        "contents": [
            {
                "parts": [
                    {
                        "text": prompt
                    }
                ]
            }
        ]
    }
    response = requests.post(url, headers=headers, json=data)  # 发送POST请求以获取API响应
    response_data = response.json()  # 将响应转换为JSON格式
    return response_data['candidates'][0]['content']['parts'][0]['text']  # 提取生成的文本响应

def save_to_md(response, filename):
    with open(filename, "w") as f:  # 打开指定文件，使用写入模式
        f.write(response)  # 将文本响应写入文件中

def open_md(filename):
    """使用Visual Studio Code打开指定的Markdown文件。

    :param filename: 要打开的Markdown文件的路径。
    """
    if os.name == "nt":  # Windows操作系统
        os.system(f"code {filename}")  # 使用code命令打开文件
    elif os.name == "posix":  # macOS或Linux操作系统
        os.system(f"open -a /Applications/Visual\\ Studio\\ Code.app {filename}")  # 使用open命令打开文件
    else:
        raise OSError("Unsupported operating system.")  # 不支持的操作系统

def generate_random_filename():
    return ''.join(random.choices(string.ascii_letters + string.digits, k=8))  # 生成随机文件名

api_key = ''  # 填写你的Gemini API密钥

api_url = ''  # 填写你的Gemini的URL

textprompt = """
# 在这里输入文本提示
"""

userprompt = """
# 在这里输入用户输入
"""

prompt = textprompt + "\n" + userprompt  # 拼接完整的提示文本

response_text = generate_response(api_key, prompt, api_url)  # 生成文本响应

# filepath = "F:\mdtmp"
## 检查目录是否存在
# if not os.path.exists(filepath):
#     # 创建目录
#     os.makedirs(filepath)

# filename = filepath + "\\" + generate_random_filename() + ".md"  # 生成保存的Markdown文件的路径和文件名
# save_to_md(response_text, filename)  # 将文本响应保存为Markdown文件
# open_md(filename)  # 使用Visual Studio Code打开保存的Markdown文件

print(response_text)  # 打印生成的文本响应
```
